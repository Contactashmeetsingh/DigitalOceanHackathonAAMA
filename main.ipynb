```python
import os
import sys
import requests

API_TOKEN = os.getenv("DO_TOKEN") or input("DigitalOcean API token: ")

headers = {"Authorization": f"Bearer {API_TOKEN}"}
response = requests.get("https://api.digitalocean.com/v2/account", headers=headers)

if response.status_code == 200:
    print("\u2705 API key is valid.")
else:
    print(f"\u274c API key test failed \u2013 HTTP {response.status_code}")
    sys.exit(1)
```
